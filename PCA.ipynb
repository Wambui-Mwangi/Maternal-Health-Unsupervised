### ***Import Statements***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### ***Loading Data***

```python
data = pd.read_csv('cleaned maternal health dataset.csv')
```

```python
drop = ['HeartRate', 'RiskLevel_Encoded']
data = data.drop(drop, axis=1)
```

```python
data.describe()
```

### ***Visualization***

```python
sns.pairplot(data)
```

### ***Scaling Features***

```python
scaling=StandardScaler()
scaled=scaling.fit_transform(data)

scaled_data=pd.DataFrame(scaled,columns=data.columns)
scaled_data.head()
```

### ***Explained Variance Plot to fond No.of Principal Components***

```python
pca = PCA().fit(scaled_data)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.show()
```

- 4 components because that's where adding components doesn't make any significant change

### ***Principal Component Analysis***

```python
components = 4
pca = PCA(n_components=components)
pca_result = pca.fit_transform(scaled_data)
```

```python
loadings_df = pd.DataFrame(pca.components_, columns=data.columns)
print(loadings_df)
```

- Most important features:
    - PC1 - SystolicBP
    - PC2 - BodyTemp
    - PC3 - Age
    - PC4 - Age
- SystolicBP, DiastolicBP, and BS have relatively high relations with the components, so they're important
- Age is important in compomnent 4 because of the strong positive loading

```python
plt.figure(figsize=(10, 6))
sns.heatmap(np.abs(loadings_df), annot=True, cmap="viridis", fmt=".2f", xticklabels=data.columns)
plt.title("Absolute Loadings Matrix")
plt.show()
```

- Blood Sugar has strong impact across all components
