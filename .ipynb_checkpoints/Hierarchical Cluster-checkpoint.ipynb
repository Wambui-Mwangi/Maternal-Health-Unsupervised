## ***Import Statements***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

## ***Loading Dataset***

```python
data = pd.read_csv('cleaned maternal health dataset.csv')
```

```python
data
```

```python
drop = ['HeartRate', 'RiskLevel_Encoded']
data = data.drop(drop, axis=1)
```

```python
data.describe()
```

## ***Data Visualization***

```python
sns.pairplot(data)
```

- We keep all features since there are visible clusters in all of them

## ***Scaling Features***

```python
scaling=StandardScaler()
scaled=scaling.fit_transform(data)
```

```python
scaled_data=pd.DataFrame(scaled,columns=data.columns)
scaled_data.head()
```

## ***Hierarchical Clustering using a Dendrogram***

```python
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")

# Perform hierarchical clustering and create a dendrogram
clusters = shc.linkage(scaled_data, method='ward', metric='euclidean')
shc.dendrogram(clusters)
plt.axhline(y=12, color='r', linestyle='--')
plt.show()
```

- Drew the horizontal line where the distance between clustering starts to get big
- Total of 13 clusters

## ***Calculting distance between clusters and linking them using Euclidean distance and Ward Linkage***

```python
clusters = 6
clustering_model = AgglomerativeClustering(n_clusters=clusters, linkage='ward', metric='euclidean')
cluster_labels = clustering_model.fit_predict(scaled_data)
```

## ***Cluster Characteristics Analysis***

```python
data['Cluster'] = cluster_labels

cluster_means = data.groupby('Cluster').mean()
print(cluster_means)
```

- Normal blood pressure - systolic<120 and diastolic<80

- Cluster 0 - middle aged women with normal blood pressure, blood sugar is the average compared to others, normal body temperature
- Cluster 1 - very young women with high blood pressure, average blood sugar level and normal body temperature
- Cluster 2 - middle aged women with high blood pressure, high blood sigar levels and normal body temperature
- Cluster 3 - very young women with normal blood pressure, average blood sugar level and normal body temperature
- Cluster 4 - young women with normal blood pressure, average blood sugar level and high body temperature
- Cluster 5 - very young women with normal blood pressure, slighlty higher blood sugar level and high body temperature
